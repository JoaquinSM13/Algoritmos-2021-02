**RESPUESTA:**

Herencia pública: Se refiere a la herencia en la que todos los miembros públicos y protegidos de la clase base conservan esos mismos niveles de acceso respectivamente en las clases derivadas.

Herencia protegida: Se refiere a la herencia en la que todos los miembros públicos de la clase base adquieren el nivel de acceso protegido en las clases derivadas, mientras que los miembros protegidos conservan su nivel de acceso.

A continuación un ejemplo de estas definiciones.

```python
%%writefile pregunta2.cpp

class Base 
{
public:
    int x;
protected:
    int y;
};

class B : public Base
{
    // x es publica
    // y es protegida
   
};

class C : protected Base
{
    // x es protegida
    // y es protegida
 
};
```

```python
%%writefile pregunta3.cpp
#include<iostream>
class Animal {

public:
    Animal();
    void Pelaje();
    ~Animal();
};
Animal::Animal(){}
void Animal::Pelaje() {//CLASE ANIMAL
    cout << "Pelaje de un animal cualquiera";
}
Animal::~Animal() {}

class Perro :public Animal {//CLASE BASE PERO QUE SOBREESCRIBE PELAJE DE LA CLASE BASE ANIMAL
public:
    Perro();
    void Pelaje();
    ~Perro();
};
Perro::Perro(){}
void Perro::Pelaje() {
    cout << "Pelaje de un perro";
}
Perro::~Perro() {}

class Buldog :public Perro {//HEREDA DE CLASE ANIMAL EL METODO PELAJE
public:
    Buldog();
    ~Buldog();
}; 
Buldog::Buldog() {}
Buldog::~Buldog() {}
int main() {//SI EJECUTAMOS PODEMOS COMPROBAR QUE SI SE PUEDE USAR UN METODO SOBREESCRITO DE UNA CLASE BASE EN ESTE CASO ANIMAL
    Perro* n1 = new Perro(); 
    Animal* n2 = new Animal(); 
    Buldog* n3 = new Buldog(); 
    n1->Pelaje(); cout << endl;
    n2->Pelaje(); cout << endl;
    n3->Pelaje(); cout << endl;
}
```

```python
%%writefile pregunta8.cpp

#include <iostream>
using namespace std;

struct Complex {
   Complex( double r, double i ) : re(r), im(i) {}
   Complex operator+( Complex &other );
   void Display( ) {   cout << re << ", " << im << endl; }
private:
   double re, im;
};

// Operator overloaded using a member function
Complex Complex::operator+( Complex &other ) {
   return Complex( re + other.re, im + other.im );
}

int main() {
   Complex a = Complex( 1.2, 3.4 );
   Complex b = Complex( 5.6, 7.8 );
   Complex c = Complex( 0.0, 0.0 );

   c = a + b;
   c.Display();
}
```

```python
%%writefile pregunta10.cpp
#include<iostream>
using namespace std;

class ClaseBase
{
    protected:
    int unavariable = 0;
    public:
    void unMetodo(void)
    {
        unavariable++;
        cout<<"Es = "<<unavariable<<endl;
    }
};

class ClaseDerivada : public ClaseBase  /* Sintaxis para indicar que ClaseDerivada hereda de ClaseBase */
{
    /* Esta clase implementa los miembros de clase que hereda de ClaseBase */
    void unMetodo(void) // Se repite el método de la clase base
    {
        unavariable++;
        cout<<"No es = "<<unavariable<<endl;
    }
};

int main()
{
    ClaseDerivada obj1;
    obj1.unMetodo(); /* Se ejecuta el método de la clase derivada */
    obj1.unMetodo();
    return 0;
}
```
